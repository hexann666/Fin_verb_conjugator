```python
# read the verbs from excel sheet

import pandas as pd

verb = pd.read_excel('Finnish_verbs.xlsx', index_col=0, header=None)
verb
```

```python
# Generate a person and a verb to conjugate

from random import randint

# dictionary for the sides of the dice and person
p = {1: 'minä', 2: 'sinä', 3: 'hän', 4: 'me', 5: 'te', 6: 'he'}

#roll the dice
dice = randint(1,6)
print(p[dice])

# choose the verb from excel-file
v = randint(1, len(verb))
print(verb.iloc[0, v])
```

```python
# Check, if the verb is familiar

transl = 'y'
if transl == ('y' or 'yes') not in input('Do you know the verb?\n').lower():
    print('{} means {}'.format(verb.iloc[0, v], verb.iloc[1, v]))
```

```python
# type in the answer

answer = input('Please type in only your answer:\n')

#check the answer
if answer == verb.iloc[1 + dice, v]:
    print('Correct')
else:
    print('Sorry, correct answer is {}'.format(verb.iloc[1 + dice, v]))
```
